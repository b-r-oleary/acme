```python
%matplotlib inline
import sys
sys.path.append('./code')

from sympy_plotting import *
from sympy import *
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_context('talk')
sns.set_style('white')

init_printing()
```

```python
D, g, W = symbols('\Delta, \gamma, \Omega')
```

```python
kappa = lambda p: sqrt(1 + p * D/sqrt(D**2 + W**2))/sqrt(2)
```

```python
e = kappa(1)
e.subs([(D,D + 1j * g)])
```

```python
manipulate([kappa(1), kappa(2)],(D, -1, 1),vary=[(W,-5,5,.1)])
```
