# Optimum Bin Size for Calculating the Mean Asymmetry

When we calculate the asymmetry, $\mathcal{A}(t)$, we typically average over $N$ adjacent data points, and within that bin we approximate the asymmetry as being a constant. We could do this in other ways such as fitting a polynomial to $\mathcal{A}(t)$, or fitting lines within a given bin, or performing local linear regression as with a non-parametric kernel method, but it is nice to choose as simple of a method as possible for interpretability.

Ultimately, consider that we have two measurements $\mathcal{A}_1(t)$ and $\mathcal{A}_2(t)$ and both are distributed in the same way. For now, lets ignore the non-gaussian tails and assume that we have sufficiently clipped the tails and the probability distribution is:

$$P(\mathcal{A}_i(t))=\mathcal{N}\left(\mu(t)=\alpha t;~\sigma^2(t)=\sigma^2_0+\frac{1}{2}\beta t^2\right)$$

such that the mean varies linearly in time and, and the variance varies quadratically in time. This is a pretty good approximation to our data:

- the signal size varies appoximately quadratically in time in the center of the pulse. The variance in the signal size S varies approximately quadratically across the center of the beam pulse, $S_i(t) \approx S_0 - (1/2)\gamma t^2$. The signals are Poissonianly distributed and hence have variance proportional to the signal size, $(\delta S_i(t))^2\propto S_i(t)$. Hence the asymmetry has variance that varies like:

$$(\delta\mathcal{A}(t))^2\propto\frac{1}{S_1(t)+S_2(t)}\propto\frac{1}{S_0}\left(1+\frac{\gamma}{2S_0}t^2\right)$$

There are several competing effects here:

- Assuming that the mean $\langle\mathcal{A}(t)\rangle=\langle\mathcal{A}\rangle$ is a constant as a function of time, then we would like to make the bin size as large as possible to minimize the variance on the mean $\langle\mathcal{A}\rangle$
- However, the probability distribution starts to become signicantly non-gaussian

```python
# create some functions for generating random data, data binning, and plotting the binned data

def gen_normal_data(t, mean=0, variance=1):
    """generate a list of random data with varying means and variances,
     where the means and variances can be arrays or scalars.
     """
    sample_data = np.random.normal(size=t.shape) * np.sqrt(variance) + mean
    return sample_data

def truncate_and_reshape(sample_data,bin_number=20):
    """ if the length of the dataset is not divisible by the bin number,
    then truncate the dataset"""
    n_bins = np.floor(len(sample_data) / bin_number) # number of bins
    truncated_length = bin_number * n_bins           # truncated length of the dataset
    
    sample_data = sample_data[0:truncated_length]            # truncating the dataset
    sample_data = sample_data.reshape((n_bins, bin_number)).T # reshaping the dataset
    return sample_data

def uniform_binning(sample_data, t=None, bin_number=20):
    'bin the data and obtain the means and variances for each bin in a list'
    sample_data = truncate_and_reshape(sample_data, bin_number=bin_number)
    
    x, dx = unweighted_mean(sample_data, axis=0)

    if t is None:
        return x, dx
    else:
        t = truncate_and_reshape(t, bin_number=bin_number)
        t = np.mean(t, axis=0)
        return t, x, dx
    
def binned_plot(t=None,sample_data=None,mean=None,std=None,bin_number=None):
    'plot the artificial data'

    plt.figure()
    if (t is not None) and (mean is not None) and (std is not None):
        #display the ideal distribution of the data
        lower    =   mean - std
        upper    =   mean + std
        
        plt.fill_between(t,lower,upper,alpha=.3)   # show the expected variance 
        
        if (bin_number is not None):
            lowern   =   mean - std/np.sqrt(bin_number)
            uppern   =   mean + std/np.sqrt(bin_number)
            plt.fill_between(t,lowern,uppern,alpha=.3) # show the expected variance in the mean
            
        plt.plot(t,mean,'--b')                     # show the mean
        
    if (t is not None) and (sample_data is not None):
        plt.plot(t,sample_data,'.r',alpha=.4)      # show the data
        
        if (bin_number is not None):
            st, smean, sstd = uniform_binning(sample_data, t=t, bin_number=bin_number)
            plt.errorbar(st,smean,yerr=sstd,alpha=.7, 
                         fmt='.b', markersize=12, capthick=2) # show the binned data
            
    plt.xlabel('time (ms)')
    plt.ylabel('$\mathcal{A}(t)$')
    if (std is not None) and (mean is not None):
        spread = (max(upper) - min(lower))/2
        center = (max(upper) + min(lower))/2

    elif (sample_data is not None):
        spread = np.std(sample_data)
        center = np.mean(sample_data)
        
    ylim=[(center - 1.5*spread), (center + 1.5*spread)]
    _ = plt.axis([min(t),max(t),ylim[0],ylim[1]])
```

## visualization of some artificial asymmetry data

first lets take a look at the form of the artificial dadta that we are working with.

```python
# set parameters for the artificial data

num_points = 300 # number of points to plot
bin_number = 20  # number of points in an averaging bin

t = np.linspace(-1,1,num_points) # time points

alpha    =   -.3 # mean slope
sigma_0  =   .02 # minimum variance
beta     =    10 # time variation of the variance

mean     =   alpha * t
variance =   sigma_0 * (1 + (1/2.0)*beta*t**2)

sample_data = gen_normal_data(t,mean,variance)

# plot the artificial data:
binned_plot(t=t,sample_data=sample_data,mean=mean,std=np.sqrt(variance),bin_number=bin_number)
```

# Simpler Problem: Unweighted Averaging $\rightarrow$ Weighted Averaging

The asymmetry problem is fairly complicated, so I am going to consider a simpler case first. In this case

We will form the following experiment:

- We split up N data-points into bins, and we varied the number of points per bin. 
- Within a bin, we computed the `unweighted_mean` and found the sample mean and the sample variance in the mean
- Outside of a bin, we computed the `weighted_mean` and found the sample mean and the sample variance
- we did this a number of times given by `reps`, and found statistics on the resulting sample mean and sample variance

```python
# create a function that provides all divisors for an input integer
def fac(n):
    """
    obtains all prime factors using code obtained from:
    http://rosettacode.org/wiki/Prime_decomposition#Python
    for the input integer n
    """
    step = lambda x: 1 + (x<<2) - ((x>>1)<<1)
    maxq = int(np.floor(np.sqrt(n)))
    d = 1
    q = n % 2 == 0 and 2 or 3 
    while q <= maxq and n % q != 0:
        q = step(d)
        d += 1
    return q <= maxq and [q] + fac(n//q) or [n]

def divisors(n):
    """
    gets all divisors of an integer n
    """
    divs = []
    divs.append(1)
    factors = fac(n) # get all prime factors
    for i in range(len(factors)+1):
        #get all combinations of length i of the factors list
        combos = itertools.combinations(factors, i) 
        for c in combos:
            #append the product of the combinations
            divs.append(np.prod(c))
    # elliminate repeats, and sort the list:
    divs = sorted(list(set(divs)))
    return divs
```

```python
def binning_experiment(N=7500, reps=1000, op1 = unweighted_mean, op2=weighted_mean):
    divs = divisors(N) # ways to divide up the data between unweighted and weighted means

    # expected sample variance:
    dy2 = 1/(N - 1)
    # expected variance in the sample variance
    ddy2 = (1/N)*np.sqrt(2/(N-1))

    W  = []
    dW = []
    V  = []
    dV = []
    Chi2 = []
    dChi2= []

    for div in divs:

        n_bin = div
        n_avg = N//div

        Z  = []
        dZ = []

        for i in range(reps):
            x = np.random.normal(size=(n_bin, n_avg))
            y, dy = op1(x, axis=0)
            z, dz = op2(y, dy)

            Z.append(z)
            dZ.append(dz)

        Z = np.array(Z)
        dZ = np.array(dZ)

        W.append(np.mean(Z))
        dW.append(np.std(Z))
        V.append(np.mean(dZ**2))
        dV.append(np.std(dZ**2))
        chi2, dchi2 = chi_squared(Z, dZ)
        Chi2.append(chi2)
        dChi2.append(dchi2)

    W = np.array(W)
    dW = np.array(dW)
    V = np.array(V)
    dV = np.array(dV)
    
    return divs, W/np.sqrt(dy2), dW/np.sqrt(dy2), V/dy2, dV/dy2, Chi2, dChi2

def plot_binning_results(divs, W, dW, V, dV, Chi2, dChi2):
    plt.figure(figsize=(6,6))
    
    ax = plt.subplot(4,1,1)
    ax.errorbar(divs, W, dW)
    ax.set_xscale('log')
    ax.set_ylabel('mean_est')
    ax.set_ylim([-5,5])

    ax = plt.subplot(4,1,2)
    ax.plot(divs, dW**2)
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_ylabel('var of mean_est')

    ax = plt.subplot(4,1,3)
    ax.errorbar(divs, V, dV)
    ax.set_xscale('log')
    ax.set_ylabel('var_est')
    ax.set_ylim([0,1.25])

    ax = plt.subplot(4,1,4)
    ax.errorbar(divs, Chi2, dChi2)
    ax.set_xscale('log')
    ax.set_ylabel('$\chi^2$')
    ax.set_xlabel('points per bin')
    ax.set_ylim([0,3])

    plt.tight_layout()
    return
```

```python
N = 7500          # total number of data points:
reps = 1000        # number of times to repeat the experment

divs, smean, dsmean, var, dvar, chi2, dchi2 = \
    binning_experiment(N=N, reps=reps, op1 = unweighted_mean, op2=weighted_mean)
plot_binning_results(divs, smean, dsmean, var, dvar, chi2, dchi2)
```

This dataset seems to show that the sample mean and the variance in the mean seems to converge to its asymtotic value right around 10 points per bin. However, at this point, the sample variance is still underestimating the actual variance by 20% which leads to $\chi^2$ values around 2 $-$ distinctly different from 1.

It seems that sample variance, and hence the measured $\chi^2$ sufficiently converge within the error  at around 100 points per bin.

Note that this unwanted behavior arises because bias or variance in the sample variance leads to incorrect weights in the `weighted_mean`, which assumes that the input variances are known exactly. When the `weighted_mean` step is replaced with the `unweighted_mean` step (which performs standard gaussian error propagation of errors after the initial binning step) does not exhibit the undesired features as seen below:

```python
N = 7500          # total number of data points:
reps = 1000        # number of times to repeat the experment

divs, smean, dsmean, var, dvar, chi2, dchi2 = \
    binning_experiment(N=N, reps=reps, op1 = unweighted_mean, op2=unweighted_mean)
plot_binning_results(divs, smean, dsmean, var, dvar, chi2, dchi2)
```

# Back to the Asymmetry

So now its clear that in the experiment we must consider three stages:

1. unweighted averaging to generating variance estimates
2. unweighted averaging propogating the initial variance estimates
3. weighted averaging to combine the results.

In the actual experiment, we perform **step 1** within a given trace (roughly 20 points per bin points). Then we perform **step 2** across an entire superblock (roughly $2^8=256$ points) and then we perform **step 3** between superblocks and across time after ablation (maybe 20 points across t, and maybe 50 superblocks).

So in the worst case scenario, where we only do unweighted averaging for initial binning and within a block, that corrsponds to around 320 data points in the unweighted averaging step. This seems ok 

```python
# lets do a check on how valid these statistics are in this case:
def test_asymmetry_stats(N, bin_number):

    means = []
    stds  = []

    for i in range(N):
        sample_data0 = gen_normal_data(t,mean,variance)
        st0, smean0, sstd0 = uniform_binning(sample_data0, t=t, bin_number=bin_number)

        sample_data1 = gen_normal_data(t,mean,variance)
        st1, smean1, sstd1 = uniform_binning(sample_data1, t=t, bin_number=bin_number)

        smean = (smean0 - smean1)
        sstd  = np.sqrt((sstd0**2 + sstd1**2))

        means.append(smean)
        stds.append(sstd)

    means = np.array(means)
    stds  = np.array(stds)

    y, dy = weighted_mean(means, stds)
    chi2, dchi2 = chi_squared(means, stds)
    
    return y, dy, chi2, dchi2

N = 1000
bin_numbers = [5, 10, 12, 15, 20, 25, 30, 50, 60, 75]

y = []
dy= []

chi2  = []
dchi2 = []

for n in bin_numbers:
    y0, dy0, chi20, dchi20 = test_asymmetry_stats(N, n)
    
    y.append(y0)
    dy.append(dy0)
    chi2.append(chi20)
    dchi2.append(dchi20)
    
y = np.array(y)
dy = np.array(dy)
chi2 = np.array(chi2)
dchi2 = np.array(dchi2)
    
ddy = dy * (2/(N * len(t) - 1))**(1/4)
        
plt.figure()
plt.subplot(2,1,1)
plt.errorbar(bin_numbers, dy, ddy)
plt.xlabel('points per bin')
plt.ylabel('error in the mean')
plt.subplot(2,1,2)
plt.errorbar(bin_numbers,chi2,dchi2)
plt.xlabel('points per bin')
plt.ylabel('$\chi^2$')
```
