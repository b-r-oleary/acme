```python
%matplotlib inline
import sys
sys.path.append('./code')
sys.path.append('../Statistics/code')

from simulations import IntegrateSchrodingerEquation
from statfunctions import PolynomialFit1D
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('notebook')
import numpy as np
import pandas as pd
```

```python
logistic = lambda t : 1/(1 + np.exp(-t))
slow_digital = lambda t, t_on, t_off, dt: logistic((t - t_on)/dt) * logistic(-(t - t_off)/dt)

T = 25
N = 1000

t = np.linspace(-T, T, N)
dt = .5

ts = np.linspace(-T, T, 6 + 2)
ts = ts[1:-1]
signs = [1,1,1]

# t_on_off= {('A','B') : (ts[2],ts[3]),
#            ('B','C') : (ts[0],ts[4]),
#            ('A','C') : (ts[1],ts[5])}

t_on_off= {('A','B') : (ts[0],ts[3]),
           ('B','C') : (ts[2],ts[4]),
           ('A','C') : (ts[1],ts[5])}

Omega = 10000
omega = {k: signs[i] * Omega * slow_digital(t, v[0], v[1], dt) 
                 for i, (k, v) in enumerate(t_on_off.items())} 

energy = {'A': 0, 
          'B': 0,
          'C': 0}

psi0 = {'A':1}
```

```python
ise = IntegrateSchrodingerEquation(t,
                                   energy=energy,
                                   omega=omega,
                                   psi0=psi0)
```

```python
plt.subplot(2,1,1)
ise.plot()
plt.subplot(2,1,2)
ise.plot_omega()
plt.tight_layout()
```
